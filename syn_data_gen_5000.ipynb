```python
import pandas as pd  
from sdv.tabular import CTGAN, GaussianCopula, CopulaGAN, TVAE
```

```python
real_data = pd.read_csv('datasets/real_data.csv')
```

### CTGAN generation

```python
ctgan = CTGAN()
ctgan.fit(real_data)
```

```python
synthetic_data = ctgan.sample(5000)
synthetic_data.to_csv('synCleanData/synCleanData_ctgan_5000.csv')
synthetic_data.head()
```

### Gaussian Copula generation

```python
gauscop = GaussianCopula()
gauscop.fit(real_data)
```

```python
synthetic_data = gauscop.sample(5000)
synthetic_data.to_csv('synCleanData/synCleanData_gauscop_5000.csv')
synthetic_data.head()
```

### CopulaGAN generation

```python
copgan = CopulaGAN()
copgan.fit(real_data)
```

### TVAE generation

```python
tvae = TVAE()
tvae.fit(real_data)
```

```python
synthetic_data = tvae.sample(5000)
synthetic_data.to_csv('synCleanData/synCleanData_tvae_5000.csv')
synthetic_data.head()
```
