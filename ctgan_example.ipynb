### CTGAN example

```python
from sdv.demo import load_tabular_demo
```

```python
data = load_tabular_demo('student_placements')
```

```python
data.head()
```

```python
from sdv.tabular import CTGAN
```

```python
model = CTGAN()
```

```python
model.fit(data)
```

```python
new_data = model.sample()
```

```python
print(new_data.head())
print(new_data.shape, data.shape)
```

### Evaluation
