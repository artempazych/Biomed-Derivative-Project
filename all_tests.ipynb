## Full Datasets Tests

```python
real_age = real['Age']
syn1_age = syn1['Age']
syn2_age = syn2['Age']
syn3_age = syn3['Age']
syn4_age = syn4['Age']

print('Real vs Syn1 (Age Column)')
print('t-test:')
print(stat_tests.t_test(real_age, syn1_age))
print('ks-test:')
print(stat_tests.ks_test(real_age, syn1_age))
print('\n')

print('Real vs Syn2 (Age Column)')
print('t-test:')
print(stat_tests.t_test(real_age, syn2_age))
print('ks-test:')
print(stat_tests.ks_test(real_age, syn2_age))
print('\n')

print('Real vs Syn3 (Age Column)')
print('t-test:')
print(stat_tests.t_test(real_age, syn3_age))
print('ks-test:')
print(stat_tests.ks_test(real_age, syn3_age))
print('\n')

print('Real vs Syn4 (Age Column)')
print('t-test:')
print(stat_tests.t_test(real_age, syn4_age))
print('ks-test:')
print(stat_tests.ks_test(real_age, syn4_age))
print('\n')
```

## Dataset Portions T-test

```python
syn1_age_ttests = []

for i in range(0, len(syn1_age), 100):
    syn1_age_ttests.append(stat_tests.t_test(real_age, syn1_age[i:i+100]))


AR_syn1_age = []
acc = 0
rej = 0
for t_test, crit, _ in syn1_age_ttests:
    if abs(t_test) > crit:
        rej+=1
        AR_syn1_age.append('Reject')
    else:
        acc+=1
        AR_syn1_age.append('Accept')

print(f'Accept {acc} | Reject {rej}')
```
